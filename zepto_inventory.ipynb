```python
import pandas as pd
```

```python
df = pd.read_csv("zepto_v2.csv", encoding="latin1")
```

```python
df.head()
```

```python
# rename columns to snake_case
df.columns = (
    df.columns
    .str.replace(r'([A-Z])', r'_\1', regex=True)   # Add underscore before capitals
    .str.lower()                                   # Make lowercase
    .str.strip('_')                                # Remove leading underscore
)
```

```python
# check for null values
df.isnull().sum()
```
